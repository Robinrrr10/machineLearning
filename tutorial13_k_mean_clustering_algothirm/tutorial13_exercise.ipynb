```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.data
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

```python
input = df[['petal length (cm)', 'petal width (cm)']]
```

```python
input
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(input[['petal length (cm)']])
```

```python
input['leng'] = scaler.transform(input[['petal length (cm)']])
```

```python
scaler.fit(input[['petal width (cm)']])
input['widt'] = scaler.transform(input[['petal width (cm)']])
```

```python
plt.xlabel('leng')
plt.ylabel('widt')
plt.scatter(input['leng'], input['widt'])
```

```python
from sklearn.cluster import KMeans
see = []
for i in range(1, 10):
    km = KMeans(n_clusters=i)
    km.fit(input[['leng', 'widt']])
    see.append(km.inertia_)
```

```python
see
```

```python
plt.xlabel('k')
plt.ylabel('sum of square error')
plt.plot(range(1, 10), see)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.target.shape
```

```python
irisData = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
irisData
```

```python
irisData['target'] = iris.target
```

```python
irisData
```

```python
irisData['type_name'] = irisData.target.apply(lambda x : iris.target_names[x])
```

```python
irisData
```

```python
iris.target_names
```

```python
irisData_setosa = irisData[irisData['type_name'] == 'setosa']
irisData_setosa
```

```python
irisData_versicolor = irisData[irisData['type_name'] == 'versicolor']
irisData_virginica = irisData[irisData['type_name'] == 'virginica'] 
```

```python
plt.scatter(irisData_setosa['sepal length (cm)'], irisData_setosa['sepal width (cm)'], marker='.', color='red')
plt.scatter(irisData_versicolor['sepal length (cm)'], irisData_versicolor['sepal width (cm)'], marker='.', color='blue')
plt.scatter(irisData_virginica['sepal length (cm)'], irisData_versicolor['sepal width (cm)'], marker='.', color='green')
```

```python
plt.scatter(irisData_setosa['petal length (cm)'], irisData_setosa['petal width (cm)'], marker='+', color='red')
plt.scatter(irisData_versicolor['petal length (cm)'], irisData_versicolor['petal width (cm)'], marker='+', color='blue')
plt.scatter(irisData_virginica['petal length (cm)'], irisData_versicolor['petal width (cm)'], marker='+', color='green')
```

```python
irisData
```

```python
X = irisData.drop(['target', 'type_name'], axis='columns')
```

```python
X
```

```python
irisData
```

```python
y = irisData.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 1000)
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X[48:52])
```

```python
y[48:52]
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
```

```python
cm
```

```python
import seaborn as sn
plt.figure(figsize=(5,3))
sn.heatmap(cm, annot=True)
plt.xlabel('Truth')
plt.ylabel('Predicted')
```

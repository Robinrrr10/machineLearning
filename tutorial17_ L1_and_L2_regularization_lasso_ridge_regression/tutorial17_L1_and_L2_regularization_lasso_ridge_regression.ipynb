### Tutorial 17 : L1 and L2 Regularization | Lasso, Ridge Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('tutorial17_melbourne_housing.csv')
df
```

```python
df.shape
```

```python
df.nunique()
```

```python
columns_needed = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dfin = df[columns_needed]
dfin
```

```python
dfin.shape
```

```python
dfin.head()
```

```python
dfin.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dfin[cols_to_fill_zero] = dfin[cols_to_fill_zero].fillna(0)
```

```python
dfin.isna().sum()
```

```python
dfin['Landsize'] = dfin['Landsize'].fillna(dfin['Landsize'].mean())
dfin['BuildingArea'] = dfin['BuildingArea'].fillna(dfin['BuildingArea'].mean())
```

```python
dfin.isna().sum()
```

```python
dfin.dropna(inplace=True)
dfin.isna().sum()
```

```python
dataFrame = pd.get_dummies(dfin, drop_first=True)
dataFrame
```

```python
X = dataFrame.drop('Price', axis=1)
y = dataFrame['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X_train, y_train)
model1.score(X_test, y_test)
```

```python
model1.score(X_train, y_train)
```

#### In above score is too less and there are more difference for test and train

#### We can use below L1 Regularization method using lasso which helps to have near to same value

```python
from sklearn.linear_model import Lasso
model2 = Lasso(alpha=50, max_iter=100, tol=0.1)
model2.fit(X_train, y_train)
model2.score(X_test, y_test)
```

```python
model2.score(X_train, y_train)
```

#### We can use below L2 Regularization method using Ridge which helps to have near to same value

```python
from sklearn.linear_model import Ridge
model3 = Ridge(alpha=50, max_iter=100, tol=0.1)
model3.fit(X_train, y_train)
model3.score(X_test, y_test)
```

```python
model3.score(X_train, y_train)
```

##### For this we came to know that L1 and L2 Regularization works better when compare with normal LinearRegression

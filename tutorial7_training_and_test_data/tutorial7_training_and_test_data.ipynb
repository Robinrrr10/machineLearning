```python
X = df[['Mileage', 'Age(yrs)']]
X
```

```python
y = df['Sell Price($)']
y
```

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, train_size=0.2)                # By this way we can split based on ratio. 0.2 means 20% for test and 80% for train
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.predict(X_test)
```

```python
y_test
```

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

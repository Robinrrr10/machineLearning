```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
input = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
input
```

```python
from sklearn.model_selection import train_test_split;
X_train, X_test, y_train, y_test = train_test_split(input, iris.target, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(RandomForestClassifier(n_estimators=200), input, iris.target)
```

```python
cross_val_score(SVC(), input, iris.target)
```

```python
cross_val_score(LogisticRegression(max_iter=1000), input, iris.target)
```

```python
cross_val_score(DecisionTreeClassifier(), input, iris.target)
```

##### As pr above mostly DecisionTreeClassifier works better here according to score
##### Similar way we can use this cross val score for checking which one performs good

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('tutorial21_heart.csv')
df
```

```python
input = df.drop('HeartDisease', axis='columns')
target = df['HeartDisease']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
input['Sex'] = le.fit_transform(input['Sex']) 
```

```python
input
```

```python
input['ChestPainType'] = le.fit_transform(input['ChestPainType']) 
input['RestingECG'] = le.fit_transform(input['RestingECG']) 
input['ExerciseAngina'] = le.fit_transform(input['ExerciseAngina']) 
input['ST_Slope'] = le.fit_transform(input['ST_Slope']) 
input
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X_scaled = scalar.fit_transform(input)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
Xs_train, Xs_test, ys_train, ys_test = train_test_split(X_scaled, target, test_size=0.3)
```

```python
from sklearn.svm import SVC
svcModel1 = SVC()
svcModel1.fit(Xs_train, ys_train)
svcModel1.score(Xs_test, ys_test)
```

```python
from sklearn.ensemble import BaggingClassifier
bag1 = BaggingClassifier(
    estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
```

```python
bag1.fit(Xs_train, ys_train)
bag1.oob_score_
```

```python
bag1.score(Xs_test, ys_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtModel2 = DecisionTreeClassifier()
dtModel2.fit(Xs_train, ys_train)
dtModel2.score(Xs_test, ys_test)
```

```python
bag2 = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
```

```python
bag2.fit(Xs_train, ys_train)
bag2.oob_score_
```

```python
bag2.score(Xs_test, ys_test)
```

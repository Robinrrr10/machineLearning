```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("tutorial2_house_prices.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel("area(sq fleet)")
plt.ylabel("price(US dollar)")
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
df_new = df.drop('price', axis='columns')
df_new
```

```python
price = df.price
price
```

```python
# create linear regression
reg = linear_model.LinearRegression()
#reg.fit(df[['area']], df.price)
reg.fit(df_new, price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[5000]])
```

```python
d = pd.read_csv('tutorial2_area.csv')
d
```

```python
p = reg.predict(d)
d['price'] = p
d
```

```python
d.to_csv('tutorial2_prediction.csv', index=False)
```

```python
plt.xlabel("area(sq fleet)")
plt.ylabel("price(US dollar)")
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue') # this blue show the prediction line
```

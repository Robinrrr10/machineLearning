```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
X = pd.DataFrame(digits.data, columns=digits.feature_names)
```

```python
y = digits.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
data_model = {
    'svc': {
        'model': SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
            },
    },
    'randomForest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 10]
        }
    },
    'logicalRegression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1, 5, 10]
        }
    },
    'decisionTree': {
        'model': DecisionTreeClassifier(),
        'params': {
        }
    },
    'gausian': {
        'model': GaussianNB(),
        'params': {
        }
    },
    'multinomial': {
        'model': MultinomialNB(),
        'params': {
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
allResult = []
for model_name, model in data_model.items():
    mod = GridSearchCV(model['model'], model['params'], cv=5, return_train_score=False)
    mod.fit(digits.data, digits.target)
    allResult.append({
        'model': model_name,
        'best_score': mod.best_score_,
        'best_params': mod.best_params_
    })
```

```python
result = pd.DataFrame(allResult)
```
